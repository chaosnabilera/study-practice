```python
# Basics of Basics

import matplotlib.pyplot
import numpy

numpy.random.seed(20231211)

n = 1000
x = (numpy.random.rand(n) - 0.5) * 20    # make it [-10, 10)]
y = (2*x) + (numpy.random.randn(n) * 3)  # y = 2x + noise [0,3)

matplotlib.pyplot.scatter(x, y)
matplotlib.pyplot.show()
```

```python
# 두개를 동시에 보여주기

n = 1000
x1 = (numpy.random.rand(n) - 0.5) * 20    # make it [-10, 10)]
y1 = (2*x1) + (numpy.random.randn(n) * 3)  # y = 2x + noise [0,3)

x2 = (numpy.random.rand(n) - 0.5) * 20     # make it [-10, 10)]
y2 = (-2*x2) + (numpy.random.randn(n) * 3)  # y = -2x + noise [0,3)

matplotlib.pyplot.scatter(x1, y1, alpha=0.5, color='blue')
matplotlib.pyplot.scatter(x2, y2, alpha=0.5, color='red')

matplotlib.pyplot.show()
```
