```python
import matplotlib.pyplot
import numpy

# random seed 고정
rng = numpy.random.default_rng(20231209)

n_points = 100000
n_bins = 20

# Generate normal distribution
dist1 = rng.standard_normal(n_points)

matplotlib.pyplot.title('Histogram example')
matplotlib.pyplot.xlabel('value')
matplotlib.pyplot.ylabel('count')

matplotlib.pyplot.hist(dist1, bins=n_bins)
matplotlib.pyplot.show()
```

```python
# bins를 늘리면 그래프가 좀 더 촘촘해 진다

n_bins = 100

matplotlib.pyplot.title('Histogram example')
matplotlib.pyplot.xlabel('value')
matplotlib.pyplot.ylabel('count')
matplotlib.pyplot.hist(dist1, bins=n_bins)

matplotlib.pyplot.show()
```

```python
# xlim, ylim을 이용하여 그래프의 범위를 지정할 수 있다

matplotlib.pyplot.xlim(-10,4)
matplotlib.pyplot.ylim(0,10000)

matplotlib.pyplot.title('Histogram example')
matplotlib.pyplot.xlabel('value')
matplotlib.pyplot.ylabel('count')
matplotlib.pyplot.hist(dist1, bins=n_bins)

matplotlib.pyplot.show()
```

```python
# 동시에 여러 histogram을 한 plot에 그릴 수 있다
# base color는 b,g,r,c,m,y,k,w 중 하나를 선택할 수 있다

dist2 = rng.normal(5, 3, n_points)

matplotlib.pyplot.title('Histogram example')
matplotlib.pyplot.xlabel('value')
matplotlib.pyplot.ylabel('count')
matplotlib.pyplot.hist(dist1, bins=n_bins, color='b', alpha=0.7)
matplotlib.pyplot.hist(dist2, bins=n_bins, color='r', alpha=0.7)

matplotlib.pyplot.show()
```

```python
# 여러개를 그린다면 legend를 보여주는게 좋다

matplotlib.pyplot.title('Histogram example')
matplotlib.pyplot.xlabel('value')
matplotlib.pyplot.ylabel('count')
matplotlib.pyplot.hist(dist1, bins=n_bins, color='b', alpha=0.7, label='dist1')
matplotlib.pyplot.hist(dist2, bins=n_bins, color='r', alpha=0.7, label='dist2')
matplotlib.pyplot.legend(['distribution 1','distribution 2'])

matplotlib.pyplot.show()
```

```python
# 여러 histogram을 동시에 보여주려면 subplot을 사용하면 된다

sub_dist1 = rng.standard_normal(n_points)
sub_dist2 = 0.4 * rng.standard_normal(n_points) + 5

# sharey = 
fig, axs = matplotlib.pyplot.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
axs[0].set_xlim(-10,10)
axs[0].hist(sub_dist1, bins=n_bins)

axs[1].set_xlim(-10,10)
axs[1].hist(sub_dist2, bins=n_bins)

matplotlib.pyplot.show()
```
