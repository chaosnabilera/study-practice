```python
# 간단한 line graph 그리기

import matplotlib.pyplot
import numpy

matplotlib.pyplot.plot([1, 2, 3, 4])
matplotlib.pyplot.show()
```

```python
# x,y 축 값 입력하기

x = numpy.arange(0, 5, 0.1)
y = 2*x + 10

matplotlib.pyplot.plot(x,y)
matplotlib.pyplot.show()
```

```python
# 두가지 동시 출력

x1 = numpy.arange(0, 5, 0.1)
y1 = 2*x + 10

x2 = numpy.arange(0, 5, 0.1)
y2 = 3*x + 5

matplotlib.pyplot.plot(x1,y1,'b--')
matplotlib.pyplot.plot(x2,y2,'r:')
matplotlib.pyplot.show()
```
